```python
#import libraries and authenticate
from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default
creds, _ = default()
import pandas as pd

gc = gspread.authorize(creds)

```

```python
#get folder/drive
# Import PyDrive and associated libraries.
# This only needs to be done once per notebook.
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials

# Authenticate and create the PyDrive client.
# This only needs to be done once per notebook.
auth.authenticate_user()
gauth = GoogleAuth()
gauth.credentials = GoogleCredentials.get_application_default()
drive = GoogleDrive(gauth)
```

```python
from pyasn1_modules.rfc5208 import Name
from os import name
#create array
ListOfFrames = []

#List all GSheets file keys/id's
# List .txt files in the root.
#
# Search query reference:
# https://developers.google.com/drive/v2/web/search-parameters
# This method returns all files by default, including trashed files. If you don't want trashed files to appear in the list, use the trashed=false query parameter to filter trashed files from the results.
listed = drive.ListFile({'q': "'1V74AfuMecLOSwi6bZ0nvHKPSLopielHf' in parents and mimeType='application/vnd.google-apps.spreadsheet' and trashed=false and title contains 'P00'"}).GetList()
counter=0

# # test print how total files
print("Total Sheets: ", len(listed))

for file in listed:
  #get sheet
  worksheet = gc.open_by_key(file['id']).sheet1
  # get_all_values gives a list of rows.
  rows = worksheet.get_all_values()
  # Convert to a DataFrame and render.
  df = pd.DataFrame.from_records(rows[1:])
  # In Pandas, the dataframe provides an attribute iloc, to select a portion of the dataframe using position based indexing.
  df = df.iloc[: , 0:10]
  # Add the PO number to end of Data Frame
  df['PO'] = file['title']
  # List of list
  ListOfFrames.append(df)
  # #Test Print
  # print(ListOfFrames[counter])
  counter+=1
  # print(file['title'])
  print("Counter: ", counter)
```

```python
combinedDF = pd.concat(ListOfFrames)
# # #Test Print w/ NaN
# combinedDF
```
